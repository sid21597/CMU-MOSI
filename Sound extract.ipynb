```python
import pandas as pd
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
object = pd.read_pickle(r'C:\Users\sidha\Downloads\mosi_data.pkl')
```

```python
object['train']['labels'][1][0][0]
```

```python
id = []
labels = []
for i in range(len(object['train']['id'])):
    id.append(object['train']['id'][i][0].decode('UTF-8'))
for i in range(len(object['train']['labels'])):
    labels.append(object['train']['labels'][i][0][0])
```

```python
for j in range(len(id)):
    filename = 'D:/CMU_MOSI/Raw/Audio/WAV_16000/Segmented/%s.wav'%id[j]
    y, sr = librosa.load(filename)
    # trim silent edges
    whale_song, _ = librosa.effects.trim(y)
    #librosa.display.waveplot(whale_song, sr=sr);
    S = librosa.feature.melspectrogram(whale_song, sr=sr, n_fft=2048, hop_length=512, n_mels=128)
    S_DB = librosa.power_to_db(S, ref=np.max)
    fig = librosa.display.specshow(S_DB, sr=sr, hop_length=512,x_axis='time', y_axis='mel');
    fig.axes.get_xaxis().set_visible(False)
    fig.axes.get_yaxis().set_visible(False)
    #plt.colorbar(format='%+2.0f dB');
    plt.savefig("D:/CMUSUER/Sound/Train/%s&%.2f.jpg"%(id[j],labels[j]), bbox_inches='tight')
    """
    for i in range(3):
        augment = whale_song+ np.random.normal(0,0.0001,len(whale_song))
        S = librosa.feature.melspectrogram(augment, sr=sr, n_fft=2048, hop_length=512, n_mels=128)
        S_DB = librosa.power_to_db(S, ref=np.max)
        fig = librosa.display.specshow(S_DB, sr=sr, hop_length=512,x_axis='time', y_axis='mel');
        fig.axes.get_xaxis().set_visible(False)
        fig.axes.get_yaxis().set_visible(False)
        #plt.colorbar(format='%+2.0f dB');
        plt.savefig("D:/CMUSUER/Sound/%s&%.2f&%d.jpg"%(id[j],labels[j],i), bbox_inches='tight')
    """
        
        
    
```
