```python
import re
result = []
for i in range(len(id)):
    
    tempid = id[i]. split("_", 1)[0]
    if id[i]. split("_", 2)[1] != re.findall(r'\d+', id[i])[-1] :
        tempid = tempid+'_'+'%s'%id[i]. split("_", 2)[1]
    if id[i]. split("_", 1)[0] == 'I5y0':
        tempid = 'I5y0__X72p0'
    df = pd.read_csv(r'D:\CMU_MOSI\Raw\Transcript\csvfiles\%s.csv'%tempid)
    temploc = int(re.findall(r'\d+', id[i])[-1])
    temptext = df.iloc[temploc-2][3]
    templabel = labels[i]
    result.append([tempid+'_'+'%s'%(re.findall(r'\d+', id[i])[-1]),temptext,templabel])

    
```
