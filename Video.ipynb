```python
import pandas as pd
import numpy as np
object = pd.read_pickle(r'C:\Users\sidha\Downloads\mosi_data.pkl')
```

```python
a = object['valid']['id']
b = object ['valid']['labels']
```

```python
id = []
labels = []
for i in range(len(object['train']['id'])):
    id.append(object['train']['id'][i][0].decode('UTF-8'))
for i in range(len(object['train']['labels'])):
    labels.append(object['train']['labels'][i][0][0])
```

```python
import numpy as np
import cv2


# Opens the Video file
for j in range(len(id)):
    cap = cv2.VideoCapture('D:/CMU_MOSI/Raw/Video/Segmented/%s.mp4'%id[j])
    i=0
    while(cap.isOpened()):
        ret, frame = cap.read()
        if ret == False:
            break
        cv2.imwrite("D:/CMUSUER/Video/Train/%s&%.2f&%d.jpg" %(id[j],labels[j],i),frame)
        i+=1

    cap.release()
    cv2.destroyAllWindows()
```
